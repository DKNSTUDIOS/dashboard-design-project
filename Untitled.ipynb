```python
import pandas as pd

import plotly.express as px

import csv

class color:
    BOLD = '\033[1m'

    
#Aufgabe 3a
#Daten einlesen und in einer strukturierten Form ausgeben

with open("flowers.csv") as csvdatei:
    csv_reader_object = csv.reader(csvdatei)

    zeilennummer = 0
    for row in csv_reader_object:

        if zeilennummer == 0:
            print(f'Spaltennamen sind: {", ".join(row)}')
        else:
            print(f' \n |Name: {row[0]} \n Hoehe: {row[1]} \n Lebenserwartung: {row[2]} \n Herkunftskontinent: {row[3]} \n Farbe: {row[4]} \n Klima: {row[5]} \n Geschnittene Blume: {row[6]} \n Parfum: {row[7]} \n Medizinisch geeignet: {row[8]} \n Durchschnittliche Anzahl Blaetter: {row[9]}.')
        zeilennummer += 1

    print(f'\n Anzahl Datensätze: {zeilennummer-1}')
```
